```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ngraph = ['1000','1000']
```

### Random

```python
train = np.load(f'Data/train_Random_{ngraph[0]}_{ngraph[1]}.npy', allow_pickle=True)
val = np.load(f'Data/val_Random_{ngraph[0]}_{ngraph[1]}.npy', allow_pickle=True)
plt.rcParams['figure.figsize'] = (12,8)
```

```python
plt.plot(train, '-b', label='train loss')
plt.plot(val.item()['bf'], '-r', label='val loss bf')
plt.plot(val.item()['bfs'], 'g', label='val loss bfs')
plt.legend(loc='upper right')
plt.title(f"Sequential Reptile K=10 {ngraph[0]}-{ngraph[1]}")
plt.xlabel("epochs")
plt.show()
```
